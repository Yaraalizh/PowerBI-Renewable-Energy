```python
import pandas as pd
import numpy as np
```

```python
jobs =pd.read_csv("Jobs Data.csv")
jobs
```

```python
jobs.info()
```

```python
duplicate_rows = jobs[jobs.duplicated()]
print(f"Number of duplicate rows: {duplicate_rows.shape[0]}")
```

```python
missing_values = jobs.isnull().sum()
print(f"Missing values in each column:\n{missing_values}")
```

```python

pivot_table = jobs.pivot_table(index='Year', columns='Technology', values='Jobs', aggfunc='sum', fill_value=0)


pivot_table
```

```python
pivot_table.to_csv('Cleaned_JobsData.csv', index=False)
```
